# Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

# Read Photo

```python
img=cv2.imread("/content/Avenger_Blog_IronMan_Imagery4-ho31fqax6.jpg")
```

# Show Image using OpenCV

```python
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Display Using Matplotlib

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

# Convert to Grey Image

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(grey_img)
```

# Invert Image

```python
invert_img=cv2.bitwise_not(grey_img)
```

```python
cv2_imshow(invert_img)
```

# Blur image

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
```

```python
cv2_imshow(blur_img)
```

# Invert Blurred Image

```python
invblur_img=cv2.bitwise_not(blur_img)
```

```python
cv2_imshow(invblur_img)
```

# Sketch

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
```

# Save Sketch

```python
cv2.imwrite("save.png",sketch_img)
```

# Display sketch

```python
cv2_imshow(sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
